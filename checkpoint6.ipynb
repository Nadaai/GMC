```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\\Users\\almostashar\\Desktop\\titanic-passengers.csv",delimiter=";")
df.head()
```

```python
df["Embarked"].value_counts()
```

```python
df["Age"].fillna(df["Age"].mean(),inplace=True)
```

```python
cleanup={"Survived":{"Yes":1, "No": 0}}  
df.replace(cleanup, inplace=True)
```

```python
cleanup={"Sex":{"male":1, "female": 0}}  
df.replace(cleanup, inplace=True)
```

```python
number_of_element=len(df["Embarked"])
print(number_of_element)
print(df["Embarked"].value_counts())
df["Embarked"].fillna("S",inplace=True)
df.head()
```

```python
l=[]
for i in df["Name"]:
    ti=i.split(", ")
    ti=ti[1].split(".")
    title=ti[0]
    l.append(title)
print(l)
```

```python
df["title"]=l
```

```python

t2=  {

                    "Capt":       "Officer",

                    "Col":        "Officer",

                    "Major":      "Officer",

                      "Dr":         "Officer",

                    "Rev":        "Officer" ,

                    "Jonkheer":   "Royalty",

                    "Don":        "Royalty",

                    "Sir" :       "Royalty",

                   "Lady" :      "Royalty" ,

                  "the Countess": "Royalty",

                    "Dona":       "Royalty" ,

                    "Mme":        "Miss",

                    "Mlle":       "Miss",

                    "Miss" :      "Miss",

                    "Ms":         "Mrs",

                    "Mr" :        "Mrs",

                    "Mrs" :       "Mrs" ,

                    "Master" :    "Master" }


```

```python
df["title"]=df["title"].map(t2)
```

```python
df.head()
```

```python
df["title"].value_counts()
```

```python
one_hot=pd.get_dummies(df["title"])
df=df.drop("title",axis=1)
df=df.join(one_hot)
df.head()
```

```python
one_hot=pd.get_dummies(df["Embarked"])
df=df.drop("Embarked",axis=1)
df=df.join(one_hot)
df.head()
```

```python
df=df.drop(["Name","PassengerId","Ticket","Fare","Cabin"], axis=1 )
df.head()
```

```python
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier 
from sklearn import metrics 
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
```

```python
X = df.iloc[:,1::]
y =df["Survived"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) 
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
text_representation = tree.export_text(clf)
print(text_representation)
```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   filled=True)
```

```python
clf = DecisionTreeClassifier(criterion="gini", max_depth=5)

clf = clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
clf=RandomForestClassifier(n_estimators=100) 
clf.fit(X_train, y_train)  
y_pred=clf.predict(X_test) 
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
